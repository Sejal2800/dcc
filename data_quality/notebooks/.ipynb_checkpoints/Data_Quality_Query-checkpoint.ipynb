```python
import datetime

# Read CSV into DataFrame
df = spark.read.option("header", "true").csv("C:/Users/Dell/Documents/data_quality/data/generated_data/Data_Quality_query.csv")

# Drop the 'title' column
df1 = df.drop('title')

# Read current counter value
cursor.execute("SELECT counter FROM dbo.counter_file_id")
row = cursor.fetchone()

df1 = df1.filter(col('file_id') == row[0] - 1)

df_query = df1.withColumn(
    'query',
    when(col('functionality') == 'null', concat(col('column_name'), lit(" IS NULL")))
    .when(col('functionality') == 'like', 
          when(col('expected').cast('int').isNotNull(), 
               concat(col('column_name'), lit(" LIKE "), col('expected'))
          ).otherwise(
               concat(col('column_name'), lit(" LIKE '"), col('expected'), lit("'"))
          )
    )
    .when(col('functionality') == 'not_like', 
          when(col('expected').cast('int').isNotNull(), 
               concat(lit("["), col('column_name'), lit("]"), lit(" != "), col('expected'))
          ).otherwise(
               concat(lit("["), col('column_name'), lit("]"), lit(" ! = '"), col('expected'), lit("'"))
          )
    )
    .when(col('functionality') == 'specific_value', 
          when(col('expected').cast('int').isNotNull(), 
               concat(col('column_name'), lit(" = "), col('expected'))
          ).otherwise(
               concat(col('column_name'), lit(" = '"), col('expected'), lit("'"))
          )
    )
    .when(col('functionality') == 'start_with', concat(col('column_name'), lit(" LIKE '"), col('expected'), lit("%'")))
    .when(col('functionality') == 'is_numeric', concat(col('column_name'), lit(" LIKE '%^[0-9]+$%'")))
    .when(col('functionality') == 'is_not_numeric', concat(col('column_name'), lit(" NOT LIKE '%^[0-9]+$%'")))
    .when(col('functionality') == 'not_contain_number', concat(col('column_name'), lit(" NOT LIKE '%[0-9]%'")))
    .when(col('functionality') == 'special_character', concat(col('column_name'), lit(" LIKE '%[^a-zA-Z0-9 ]%'")))
    .when(col('functionality') == 'is_lower', concat(col('column_name'), lit(" = LOWER("), col('column_name'), lit(")")))
    .when(col('functionality') == 'fixed_length', concat(lit("LEN(CONCAT("), col('column_name'), lit(", '')) = "), col('expected')))
    .when(col('functionality') == 'ranged_length', concat(
        lit("LEN("), col("column_name"), lit(") BETWEEN "), 
        split(col("expected"), ",")[0], lit(" AND "), split(col("expected"), ",")[1]
    ))
    .when(col('functionality') == 'date_format', when(col('expected').isNotNull(), 
        concat(col('column_name'), lit(" LIKE '"), coalesce(col('expected'), lit("YYYYMMDD")), lit("'"))
    ).otherwise(lit("1 = 1")))
    .when(col('functionality') == 'length', concat(
        lit("LEN(CONCAT("), col('column_name'), lit(", '')) < "), coalesce(col('expected'), lit("10"))
    ))
    .when(trim(lower(col('functionality'))) == 'duplicity', lit('1 = 1'))
)
```

```python
# Function to construct SQL query safely
def query_table(d):
    return f"""
                SELECT DISTINCT
                    {d['file_id']} AS file_id,
                    '{d['rule_no']}' AS rule_no, 
                    '{d['business_rule_definition'].replace("'", "''") if d['query'] else ''}' AS business_rule_definition,
                    'unknown' AS customer, 
                    NULL AS account_group,
                    '{d['table_name']}' AS table_name,
                    '{d['column_name']}' AS column_name,
                    '{d['column_name']}' AS business_attribute_name,
                    '{d['attribute_group']}' AS attribute_group, 
                    'High' AS severity, 
                    NULL AS dependent_column_name,			   		        
                    NULL AS dependent_column_value, 
                    NULL AS country, 
                    '{d['functionality']}' AS Functionality, 			 
                    '{d['dimension']}' AS dimension, 
                    CAST(GETDATE() AS DATE) AS extract_date,  
                    NULL AS Cleaned_flag,
                    '{d['query'].replace("'", "''") if d['query'] else ''}' AS query
                FROM dbo.{d['table_name']} 
                WHERE {d['query']}
                """

# List to store successful queries
successful_records = []

# List to store failed queries
failed_records = []

c = 0  # Counter for failed queries

# Define the schema for successful records
schema = StructType([
    StructField("file_id", StringType(), True),
    StructField("rule_no", StringType(), True),
    StructField("business_rule_definition", StringType(), True),
    StructField("customer", StringType(), True),
    StructField("account_group", StringType(), True),
    StructField("table_name", StringType(), True),
    StructField("column_name", StringType(), True),
    StructField("business_attribute_name", StringType(), True),
    StructField("attribute_group", StringType(), True),
    StructField("severity", StringType(), True),
    StructField("dependent_column_name", StringType(), True),
    StructField("dependent_column_value", StringType(), True),
    StructField("country", StringType(), True),
    StructField("Functionality", StringType(), True),
    StructField("dimension", StringType(), True),
    StructField("extract_date", StringType(), True),
    StructField("Cleaned_flag", StringType(), True),
    StructField("query", StringType(), True)
])

failed_schema = StructType([
    StructField("file_id", StringType(), True),
    StructField("table_name", StringType(), True),
    StructField("column_name", StringType(), True),
    StructField("rule_no", StringType(), True),
    StructField("business_rule_definition", StringType(), True),
    StructField("attribute_group", StringType(), True),
    StructField("dimension", StringType(), True),
    StructField("functionality", StringType(), True),
    StructField("expected", StringType(), True),
    StructField("query", StringType(), True),
    StructField("error_message", StringType(), True)
])
```

```python
# Execute Queries
# c = 0
for row in df_query.collect():
    # c = c + 1
    sql_query = query_table(row.asDict())
    # print(c, sql_query)
    try:
        cursor.execute(sql_query)
        successful_records.append(cursor.fetchall())
    except Exception as e:
        failed_records.append((
            row["file_id"],
            row["table_name"],
            row["column_name"],
            row["rule_no"],
            row["business_rule_definition"],
            row["attribute_group"],
            row["dimension"],
            row["functionality"],
            row["expected"],
            row["query"],
            str(e)
        ))
```

```python
# Write failed records to CSV
if failed_records:
    df_failed = spark.createDataFrame(failed_records, failed_schema)
    df_failed_pd = df_failed.toPandas()
    df_failed_pd.to_sql('failed_queries', con=engine, if_exists='append', index=False)

# Write successful records to CSV
if successful_records:
    # Flatten the nested list
    flat_records = [item for sublist in successful_records for item in sublist]

    # Convert date to string format for consistency with schema
    flat_records = [
        tuple(str(value) if isinstance(value, (int, float, bool)) or value is None else value.strftime('%Y-%m-%d') if isinstance(value, datetime.date) else value
        for value in record
    ) for record in flat_records]
    
    # Create DataFrame
    df_successful = spark.createDataFrame(flat_records, schema=schema)
    df_successful_pd = df_successful.toPandas()
    df_successful_pd.to_sql('successful_queries', con=engine, if_exists='append', index=False)
```
