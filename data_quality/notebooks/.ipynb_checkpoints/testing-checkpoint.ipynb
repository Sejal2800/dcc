```python
from pyspark.sql import SparkSession

# Create SparkSession
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("Jupyter PySpark") \
    .config("spark.executor.memory", "2g") \
    .getOrCreate()    

# Verify Spark Context
# print(spark.sparkContext.getConf().getAll())
```
