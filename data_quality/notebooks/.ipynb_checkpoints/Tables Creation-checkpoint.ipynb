
## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```

from pyspark.sql import SparkSession

# Initialize SparkSession
spark = SparkSession.builder.enableHiveSupport().getOrCreate()

# Define file paths
file_paths = [
    "dbfs:/FileStore/tables/Customer.csv",
    "dbfs:/FileStore/tables/CustomerTexts.csv",
    "dbfs:/FileStore/tables/CustomerType.csv",
    "dbfs:/FileStore/tables/CustomerTypeTexts.csv",
    "dbfs:/FileStore/tables/CustomersHierarchy.csv",
    "dbfs:/FileStore/tables/FinancialTransactions.csv",
    "dbfs:/FileStore/tables/GLAccountType.csv",
    "dbfs:/FileStore/tables/GLAccountTypeTexts.csv",
    "dbfs:/FileStore/tables/GLAccountsHierarchy.csv",
    "dbfs:/FileStore/tables/GLAccountsTexts.csv",
    "dbfs:/FileStore/tables/GLAccounts.csv",
    "dbfs:/FileStore/tables/ProductCategories.csv",
    "dbfs:/FileStore/tables/ProductCategoryTexts.csv",
    "dbfs:/FileStore/tables/ProductHierarchy.csv",
    "dbfs:/FileStore/tables/ProductTexts.csv",
    "dbfs:/FileStore/tables/Product.csv",
    "dbfs:/FileStore/tables/ProfitCenter.csv",
    "dbfs:/FileStore/tables/ProfitCenterHierarchy.csv",
    "dbfs:/FileStore/tables/ProfitCenterTexts.csv"
]

# Set database (Hive metastore)
database_name = "default"

# Ensure the database exists
spark.sql(f"CREATE DATABASE IF NOT EXISTS {database_name}")

# Loop through files and create tables
for file_path in file_paths:
    table_name = file_path.split("/")[-1].replace(".csv", "").lower()  # Extract filename as table name
    
    # Read CSV file
    df = (
        spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(file_path)
    )
    
    # Register as a Hive table in the default metastore
    full_table_name = f"{database_name}.{table_name}"
    df.write.mode("overwrite").saveAsTable(full_table_name)
    
    print(f"✅ Registered table: {full_table_name}")
```

```
from pyspark.sql import SparkSession

# Initialize SparkSession with Hive support
spark = SparkSession.builder.enableHiveSupport().getOrCreate()

# Define database name (Hive metastore)
database_name = "default"

# Ensure the database exists
spark.sql(f"CREATE DATABASE IF NOT EXISTS {database_name}")

# Get all CSV files dynamically from the folder
file_paths = [file.path for file in dbutils.fs.ls("dbfs:/FileStore/tables/data") if file.name.endswith(".csv")]

# Loop through files and create tables
for file_path in file_paths:
    table_name = file_path.split("/")[-1].replace(".csv", "").lower()  # Extract filename as table name
    
    # Read CSV file with inferred schema
    df = (
        spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(file_path)
    )
    
    # Save as Hive Table
    full_table_name = f"{database_name}.{table_name}"
    df.write.mode("overwrite").saveAsTable(full_table_name)
    
    print(f"✅ Registered table: {full_table_name}")
```
