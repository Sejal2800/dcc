```python
import findspark
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.functions import expr
import pyodbc
import datetime
from sqlalchemy import create_engine
import os
from pyspark.sql import SparkSession
from pyspark.sql.types import *
import random
from datetime import datetime
import pandas as pd

findspark.init()

print(os.environ.get("HADOOP_HOME"))

os.environ['HADOOP_HOME'] = "C:\\hadoop-3.0.0"
os.environ['PATH'] += os.pathsep + 'C:\\hadoop-3.0.0\\bin'

# Create SparkSession
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("Jupyter PySpark") \
    .config("spark.executor.memory", "2g") \
    .config("spark.sql.sources.commitProtocolClass", "org.apache.spark.sql.execution.datasources.SQLHadoopMapReduceCommitProtocol") \
    .config("spark.sql.parquet.output.committer.class", "org.apache.hadoop.mapreduce.lib.output.DirectFileOutputCommitter") \
    .getOrCreate()    

# Setup connection to SQL Server
conn = pyodbc.connect(
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=localhost;'
    'DATABASE=master;'
    'Trusted_Connection=yes;'
)

cursor = conn.cursor()

# Create SQLAlchemy engine using pyodbc
engine = create_engine(
    'mssql+pyodbc://localhost/master?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes'
)
```
